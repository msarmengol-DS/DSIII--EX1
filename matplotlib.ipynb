```python
iris_df = sns.load_dataset("iris")
iris_df.head()
```

```python
num_rows = iris_df.shape[0]

sns.histplot(data=iris_df, x="sepal_length")
```

```python
sepal_length = iris_df["sepal_length"]
sepal_width = iris_df["sepal_width"]
petal_specie = iris_df["species"]

sns.scatterplot(x=sepal_length, y=sepal_width, hue=petal_specie)
```
